```python
import json
import requests
import os
import urllib
import gzip
```

```python
P114_INPUT_DIR = '/home/graeme/ElexonData/P114/'
ELEXON_LIST_URL = 'https://downloads.elexonportal.co.uk/p114/list?key={}'
ELEXON_DOWNLOAD_URL = 'https://downloads.elexonportal.co.uk/p114/download?key={}&filename={}'
ELEXON_KEY = ''
```

```python
response = requests.get(ELEXON_LIST_URL.format(ELEXON_KEY))
json_data = json.loads(response.text)

json_data
```

```python
for filename, timestamp in json_data.items():
    if not os.path.isfile(P114_INPUT_DIR + filename):
        remote_url = (ELEXON_DOWNLOAD_URL.format(ELEXON_KEY, filename))
        urllib.request.urlretrieve(
            remote_url,
            P114_INPUT_DIR + filename)
```

```python
filename = list(json_data.keys())[3]
file = gzip.open(P114_INPUT_DIR + filename, 'rb')
file_content = file.read().decode('utf-8', 'ignore')

for row in file_content.split('\n'):
    print(row)
    #print(row.split('|'))
```
