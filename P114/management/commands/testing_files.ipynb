```python
import json
import requests
import os
import urllib
import gzip
import sys
import datetime as dt
from ElexonDataManager.settings import ELEXON_KEY, P114_INPUT_DIR, P114_LIST_URL, P114_DOWNLOAD_URL
```

```python
os.environ['PYTHONPATH']
sys.path
```

```python
P114_date = dt.date(2019,1,10)
print(P114_LIST_URL.format(ELEXON_KEY, P114_date.year, P114_date.month, P114_date.day))
response = requests.get(P114_LIST_URL.format(ELEXON_KEY, P114_date.year, P114_date.month, P114_date.day))
json_data = json.loads(response.text)

json_data
```

```python
for filename, timestamp in json_data.items():
    if not os.path.isfile(P114_INPUT_DIR + filename):
        remote_url = (P114_DOWNLOAD_URL.format(ELEXON_KEY, filename))
        urllib.request.urlretrieve(
            remote_url,
            P114_INPUT_DIR + filename)
```
